```python
from scipy.io.idl import readsav
from types import SimpleNamespace
from glob import glob
import pandas as pd
import numpy as np

files=sorted(glob("/home/vampy/acads/projects/Probing_high_freq_waves_in_corona/Data/solo_L2_EUI-HRIEUV174-IMAGE_2022-10-13T13:06:00_2022-10-13T14:35:51/F1/Slits/*"))

# slit=sorted(['S1','S2','S4','S5','S6','S7','S8'])
# slit=sorted(['S3','S9','S10','S11','S12','S13'])
slit=sorted(['G1','G2','G3','G4','G5','G6'])
slit1=sorted(['S1','S2','S3','S4','S5','S6'])

k=0;

#slit_info=pd.read_csv(slit+"/"+"info.csv",sep=',',header=0,names=["slit name","xi","yi","xf","yf","width","folder"])

for i in files:
    #file_name="/home/vampy/acads/projects/Probing_high_freq_waves_in_corona/Data/solo_L2_EUI-HRIEUV174-IMAGE_2022-03-17T03:18:00_2022-03-17T04:02:00/R1/Slits/xtS1.sav"
    file_name=i+f"/xt{slit1[k]}.sav"
    print(file_name)

    df=readsav(file_name, python_dict=True, verbose=True)
    var =  SimpleNamespace(**df)
    xt=var.xt
    np.savetxt(i+"/xt_map.csv",xt.T,delimiter=',')
    print(i+"/xt_map.csv")
    k+=1
    print("------")
```
